```python
df_curso['CINE_AREA'] = df_curso['CO_CINE_ROTULO'].str[:2]
```

```python
df_curso['CINE_AREA']
```

```python
lista_ies_curso = df_curso.groupby(['CO_IES','CINE_AREA']).count().index
```

```python
dict_ies_curso = dict()
for i in lista_ies_curso:
    if i[0] in dict_ies_curso.keys():
        dict_ies_curso[i[0]].append(i[1])
    else:
        dict_ies_curso[i[0]] = [i[1]]
```

```python
dict_ies_curso
```

```python
lista_unis_ti = []
for i in dict_ies_curso:
    if len(dict_ies_curso[i]) == 1:
            print(dict_ies_curso[i])
    for j in dict_ies_curso[i]:
        if j != '06':
            # print(i)
            break
    else:
        lista_unis_ti.append(i)
```

```python
print(len(lista_unis_ti))
lista_unis_ti
```

```python
unis_ti = df_uni.loc[df_uni.CO_IES.isin(lista_unis_ti) & (df_uni.NU_ANO_CENSO >= 2015)]
```
