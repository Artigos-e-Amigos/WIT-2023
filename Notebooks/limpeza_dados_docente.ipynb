# Notebook de limpeza e junção de dados de docentes do INEP

## Imports e funções

```python
import pandas as pd
```

```python
dados_2019 = pd.read_csv('../Data/SUP_DOCENTE_2019.CSV', sep='|', encoding='ISO-8859-1')
dados_2018 = pd.read_csv('../Data/DM_DOCENTE.CSV', sep='|', encoding='ISO-8859-1')
dados_2017 = pd.read_csv('../Data/SUP_DOCENTE_2017.CSV', sep='|', encoding='ISO-8859-1')
dados_2016 = pd.read_csv('../Data/SUP_DOCENTE_2016.CSV', sep='|', encoding='ISO-8859-1')
dados_2015 = pd.read_csv('../Data/SUP_DOCENTE_2015.CSV', sep='|', encoding='ISO-8859-1')
```

## Juntando as bases

```python
bases = [dados_2015, dados_2016, dados_2017, dados_2018, dados_2019]
soma_tamanho_bases = 0
for b in bases:
    soma_tamanho_bases += len(b)

print(f'Soma do tamanho das bases: {soma_tamanho_bases}')
```

```python
dados_2015_2019 = pd.concat(bases)
dados_2015_2019.info()
```

## Salvando a nova base

```python
dados_2015_2019.to_csv('../Data/docentes_2015-2019.csv', index=False)
```
